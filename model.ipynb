```python
def estimate_path(action_dict, threshold = 0.36):
    path = [action for action, freq in action_dict.items() if freq > threshold]
    path.sort(key = lambda a: action_dict[a], reverse = True)
    
    return path
```

```python
def simulate(data,func, previos = [[],[],[]]):
    if func == AutoPath or func == AutoPathReal:
        X = Xauto
    elif func == TelePath or func == TelePathReal:
        X = Xtele
    else:
        return 0
    team1 = estimate_path(func(list(data.keys())[0],data[list(data.keys())[0]]), X)
    team2 = estimate_path(func(list(data.keys())[1],data[list(data.keys())[1]]), X)
    team3 = estimate_path(func(list(data.keys())[2],data[list(data.keys())[2]]), X)
    
    if func == AutoPath or func == AutoPathReal:
        team1 = team1[:5]
        team2 = team2[:5]
        team3 = team3[:5]
    
    if set(team1) & set(team2):
        commons = set(team1).intersection(team2)
        for i in commons:
            T1 = (func(list(data.keys())[0],data[list(data.keys())[0]])[i]) + 0.05
            T2 = func(list(data.keys())[1],data[list(data.keys())[1]])[i]
            if (len(team1) > 5) & (len(team1) != len(team2)):
                team1.remove(i)
            elif (len(team2) > 5) & (len(team1) != len(team2)):
                team2.remove(i)
            else:
                if T2>T1:
                    team1.remove(i)
                else:
                    team2.remove(i)
    
    if set(team2) & set(team3):
        commons = set(team2).intersection(team3)
        for i in commons:
            T3 = (func(list(data.keys())[2],data[list(data.keys())[2]])[i]) + 0.05
            T2 = func(list(data.keys())[1],data[list(data.keys())[1]])[i]
            if T2>T3:
                team3.remove(i)
            else:
                team2.remove(i)
    
    if set(team1) & set(team3):
        commons = set(team1).intersection(team3)
        for i in commons:
            T1 = func(list(data.keys())[0],data[list(data.keys())[0]])[i]
            T3 = func(list(data.keys())[2],data[list(data.keys())[2]])[i]
            if T3>T1:
                team1.remove(i)
            else:
                team3.remove(i)
    # previos[0] = {l: globals()[f"autoReef_{l}"] for l in previos[0]}
    # previos[1] = {l: globals()[f"autoReef_{l}"] for l in previos[1]}
    # previos[2] = {l: globals()[f"autoReef_{l}"] for l in previos[2]}
    TM1 = team1
    TM2 = team2
    TM3 = team3
    
    if func == TelePath or func == TelePathReal:
        TM1 = [name.replace('teleop', 'auto') for name in team1]
        TM2 = [name.replace('teleop', 'auto') for name in team2]
        TM3 = [name.replace('teleop', 'auto') for name in team3]
    
    TMS = TM1 + TM2 +TM3
    full_previos = previos[0] + previos[1] + previos[2]
    
    if set(TM1) & set(full_previos):
        # TM1 = TM1.remove(set(TM1).intersection(full_previos))
        TM1 = list(set(TM1) - set(full_previos))
    if set(TM2) & set(full_previos):
        # TM2 = TM2.remove(set(TM2).intersection(full_previos))
        TM2 = list(set(TM2) - set(full_previos))
    if set(TM3) & set(full_previos):
        # TM3 = TM3.remove(set(TM3).intersection(full_previos))
        TM2 = list(set(TM2) - set(full_previos))
        
    return [TM1,TM2,TM3]
```
