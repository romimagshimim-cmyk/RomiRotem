```python
df['blue_Robot1'] = df['blue_teams_lst'].apply(lambda x: x[0] if len(x) > 0 else None)
df['blue_Robot2'] = df['blue_teams_lst'].apply(lambda x: x[1] if len(x) > 0 else None)
df['blue_Robot3'] = df['blue_teams_lst'].apply(lambda x: x[2] if len(x) > 0 else None)
```

```python
df['red_Robot1'] = df['red_teams_lst'].apply(lambda x: x[0] if len(x) > 0 else None)
df['red_Robot2'] = df['red_teams_lst'].apply(lambda x: x[1] if len(x) > 0 else None)
df['red_Robot3'] = df['red_teams_lst'].apply(lambda x: x[2] if len(x) > 0 else None)
```

```python
df.drop(['blue_teams_lst', 'red_teams_lst', 'blue_teams', 'red_teams'], axis = 1, inplace = True)
```

```python
df.drop(['red_totalPoints', 'blue_totalPoints'], axis=1, inplace=True)
```
