```python
import pandas as pd
import time
import re
```

```python
path = r'C:\Users\Magsihim_AI\Desktop\Project_Statbotics\full_table2.csv'

df = pd.read_csv(path)
```

```python
df.drop(['actual_time', 'key', 'post_result_time', 'time', 'videos', 'adjustPoints'], axis=1, inplace=True)
```

```python
df
```

```python
#change
#endGame, team_number, alliance, comp_level, event_key, winning_alliance, autoLine
```

```python
df['endGame'].unique()
```

```python
df['endGame'] = df['endGame'].map({'Nothing': 0, 'Parked': 1, 'ShallowCage': 2, 'DeepCage': 3})
```

```python
df['red'] = df['alliance'] == 'red'
df['blue'] = df['alliance'] == 'blue'
```

```python
df['winning_red'] = df['winning_alliance'] == 'red'
df['winning_blue'] = df['winning_alliance'] == 'blue'
```

```python
df.drop(['alliance', 'winning_alliance'], axis = 1, inplace = True)
```

```python
# df['team_number'] = df['team_number'].map(lambda x: int(re.search(r'\d+', x).group()))
```

```python
df['team_number'] = df['team_number'].map(
    lambda x: int(re.search(r'\d+', str(x)).group() + ("1" if str(x).strip().endswith("B") else ""))
)
```

```python
df['comp_level'].unique()
```

```python
df['comp_level'] = df['comp_level'].map({'qm': 0, 'sf': 1, 'f': 2})
```

```python
# df['event_key'] = pd.factorize(df['event_key'])[0]
```

```python
df['autoLine'] = df['autoLine'].map({'Yes': True, 'No': False})
```

```python
df
```

```python
df_info = pd.DataFrame({
    'column': df.columns,
    'non_nulls': df.notnull().sum(),
    'dtype': df.dtypes
}).reset_index(drop=True)
```

```python
df_info['dtype'].unique()
```

```python
df['win'] = ((df['blue'] & df['winning_blue']) | (df['red'] & df['winning_red']))
```
