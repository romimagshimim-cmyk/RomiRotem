```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = r'C:\Users\Magsihim_AI\Desktop\Project_Statbotics\data\datasets\final_table.csv'

df = pd.read_csv(path)
```

```python
drops = ['comp_level', 'event_key', 'match_number', 'set_number', 'algaePoints', 'autoBonusAchieved', 'autoCoralCount', 'autoCoralPoints', 'autoMobilityPoints', 'autoPoints', 'bargeBonusAchieved', 'coopertitionCriteriaMet', 'coralBonusAchieved', 'foulPoints', 'g410Penalty', 'g418Penalty', 'g428Penalty', 'rp', 'teleopCoralCount', 'teleopCoralPoints', 'teleopPoints', 'scorePenaltyFree', 'red', 'blue', 'winning_red', 'winning_blue', 'endGameBargePoints']
```

```python
df.drop(drops, axis = 1, inplace = True)
```

```python
df.drop(['teleopReef_tba_botRowCount', 'teleopReef_tba_midRowCount', 'teleopReef_tba_topRowCount', 'autoReef_tba_botRowCount', 'autoReef_tba_midRowCount', 'autoReef_tba_topRowCount'], axis = 1, inplace = True)
```

```python
df.to_csv(r"C:\Users\Magsihim_AI\Desktop\Project_Statbotics\actions_data\datasets\actions_df.csv", index=False)
```

```python
# correlations with target
correlations = df.corr(numeric_only=True)["win"].drop("win")  # drop self-correlation
corr_df = correlations.to_frame().reset_index()
corr_df.columns = ["Feature", "Correlation"]

# sort by absolute correlation
corr_df = corr_df.reindex(corr_df["Correlation"].abs().sort_values(ascending=False).index)

plt.figure(figsize=(8, len(corr_df) * 0.3))
sns.barplot(
    data=corr_df,
    y="Feature", x="Correlation",
    palette="coolwarm"
)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Feature Correlations with Target", fontsize=14)
plt.tight_layout()
plt.show()
```
