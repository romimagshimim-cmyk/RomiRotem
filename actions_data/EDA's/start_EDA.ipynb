```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
path = r'C:\Users\Magsihim_AI\Desktop\Project_Statbotics\actions_data\datasets\actions_df.csv'

df = pd.read_csv(path)
```

```python
# correlations with target
correlations = df.corr(numeric_only=True)["win"].drop("win")  # drop self-correlation
corr_df = correlations.to_frame().reset_index()
corr_df.columns = ["Feature", "Correlation"]

# sort by absolute correlation
corr_df = corr_df.reindex(corr_df["Correlation"].abs().sort_values(ascending=False).index)

plt.figure(figsize=(8, len(corr_df) * 0.3))
sns.barplot(
    data=corr_df,
    y="Feature", x="Correlation",
    palette="coolwarm"
)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Feature Correlations with Target", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# correlations with target
correlations = df.corr(numeric_only=True)["score"].drop("score")  # drop self-correlation
corr_df = correlations.to_frame().reset_index()
corr_df.columns = ["Feature", "Correlation"]

# sort by absolute correlation
corr_df = corr_df.reindex(corr_df["Correlation"].abs().sort_values(ascending=False).index)

plt.figure(figsize=(8, len(corr_df) * 0.3))
sns.barplot(
    data=corr_df,
    y="Feature", x="Correlation",
    palette="coolwarm"
)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Feature Correlations with Target", fontsize=14)
plt.tight_layout()
plt.show()
```
